Heart Disease Detection using multivariate linear regression

```python
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
def gradient(X, y, w1, b1): 
    m,n=X.shape
     
    dj_dw = np.zeros((n,))
    dj_db = 0.

    for i in range(m):                             
        err = (np.dot(X[i], w1) + b1) - y[i]   
        for j in range(n):                         
            dj_dw[j] = dj_dw[j] + err * X[i, j]    
        dj_db = dj_db + err                        
    dj_dw = dj_dw / m                                
    dj_db = dj_db / m                                
        
    return dj_db, dj_dw
```

```python
def gradient_descent(X, y,w2,b2,alpha, num_iters): 
    for i in range(num_iters):
        dj_db,dj_dw = gradient(X, y, w2, b2)   
        w2 = w2 - alpha * dj_dw               
        b2 = b2 - alpha * dj_db                   
        if(i%100==0):
            print("Cost after iteration ",i)
            print(cost(X,y,w2,b2))
    return w2, b2
```

```python
def cost(X, y, w3, b3): 
    m = X.shape[0]
    cost = 0.0
    for i in range(m):                                
        f_wb_i = np.dot(X[i], w3) + b3          
        cost += (f_wb_i - y[i])**2       
    cost = cost / (2 * m)                        
    return cost
```

```python
df=pd.read_csv("heart.csv")
X=df.loc[:,df.columns!='heart.disease'].values[:,1:]
y=np.array(df["heart.disease"])

print(X.shape)
print(y.shape)
```

```python
w=np.array([1,1.5])
b=2
w,b=gradient_descent(X,y,w,b,0.0009,10000)
```
